```python
selected_features=pd.concat([selected_features, disney_features, amazonprime_features], axis=0)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
  stemmed_text = [ps.stem(word) for word in text.split()]
  return " ".join(stemmed_text)
```

```python
selected_features['combined_features'] = selected_features['title']+' '+selected_features['type']+' '+selected_features['director']+' '+selected_features['cast']+' '+selected_features['listed_in']+' '+selected_features['description']
```

```python
selected_features.shape
```

```python
selected_features['index'] = range(1, len(selected_features) + 1)
```

```python
selected_features.head()
```

```python
selected_features['combined_features'] = selected_features['combined_features'].apply(stem)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = list(stopwords.words('english'))
feature_extraction = TfidfVectorizer(analyzer = 'word', stop_words=stop, ngram_range=(1,2), max_features=10000)
```

```python
feature_vectors = feature_extraction.fit_transform(selected_features['combined_features'])
```

```python
print(feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
similarity.shape
```

Finding Matches

```python
movie_name = input('Enter Movie Name:')
list_of_all_titles = selected_features['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, selected_features['title'])
print(find_close_match)

# Get the index using .iloc to avoid potential issues
index_of_the_movie = selected_features[selected_features['title'] == find_close_match[0]].index.values[0]
print(index_of_the_movie)

similarity_score = list(enumerate(similarity[index_of_the_movie]))
sorted_similar_movies = sorted(similarity_score, key = operator.itemgetter(1), reverse= True)
print ('Suggested Movies And Show : \n ')
i=1
for movies in sorted_similar_movies:
  index = movies[0]
  # Use .iloc to access data by position instead of the index label
  title_from_index = selected_features.iloc[[index]]['title'].values[0]
  channel_from_index = selected_features.iloc[[index]]['channel'].values[0]
  if (i<30):
    print(i, '.',title_from_index,channel_from_index)
    i+=1
```
